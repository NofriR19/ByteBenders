### 2. Univariate Analysis (25 poin)

Gunakan visualisasi untuk melihat distribusi masing-masing kolom (feature maupun
target). Tuliskan hasil observasinya, misalnya jika ada suatu kolom yang distribusinya
menarik (misal skewed, bimodal, ada outlier, ada nilai yang mendominasi, kategorinya
terlalu banyak, dsb). Jelaskan juga apa yang harus di-follow up saat data pre-processing.

```python
features = nums 
plt.figure(figsize=(30, 40))

# Menentukan jumlah baris dan kolom berdasarkan jumlah fitur
total_features = len(features)
rows = (total_features // 4) + (total_features % 4 > 0)
cols = min(4, total_features)

# Membuat subplot grid
for i in range(total_features):
    plt.subplot(rows, cols, i + 1)
    sns.kdeplot(x=df[features[i]])
    plt.xlabel(features[i])
    plt.title(f'{features[i]}')  # Menambahkan judul untuk setiap subplot

# Menambahkan judul keseluruhan plot
plt.suptitle('Distribusi kolom numerikal', fontsize=20)

# Menyesuaikan layout
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='NumberOfPersonVisiting', data=df)
plt.title('Barplot untuk Kolom NumberOfPersonVisiting')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='NumberOfFollowups', data=df)
plt.title('NumberOfFollowups')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='PreferredPropertyStar', data=df)
plt.title('PreferredPropertyStar')
plt.show()
```

```python
numss = ["MonthlyIncome"]

fig, axes = plt.subplots(1, len(numss), figsize=(25, 10))

for i, num in enumerate(numss):
    sns.histplot(df[num], kde=True, ax=axes)
    axes.set_title(f'Monthly Income')
    axes.set_xlabel(num)

plt.tight_layout()
plt.show()
```

```python
prod_taken_counts = df['ProdTaken'].value_counts()

plt.figure(figsize=(10, 10))
plt.pie(prod_taken_counts, labels=prod_taken_counts.index, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightcoral'])
plt.title('ProdTaken')
plt.show()
```

```python
categorical_columns = ['TypeofContact', 'Occupation', 'Gender', 'ProductPitched', 'MaritalStatus', 'Designation']

# Membuat subplot grid
fig, axes = plt.subplots(2, 3, figsize=(20, 25))

# Iterasi untuk setiap kolom kategorikal
for col, ax in zip(categorical_columns, axes.flatten()):
    # Membuat countplot untuk setiap kolom
    sns.countplot(x=col, data=df, ax=ax)
    ax.set_title(f'{col}')

# Menyesuaikan layout
plt.tight_layout()
plt.show()
```
